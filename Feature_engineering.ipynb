###Feature Engineering
Feature engineering is the process of transforming raw data into meaningful features that better represent the underlying problem to the machine learning model. It involves creating new features or modifying existing ones to improve the modelâ€™s performance.

Feature engineering can often lead to significant improvements in a model's performance, and it's an essential step in the machine learning pipeline.

```python
from sklearn.preprocessing import PolynomialFeatures

# Example feature
X = [[2], [3], [4]]

# Creating a polynomial feature (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

print(X_poly)
```

```python
import numpy as np

# Example feature
X = np.array([1, 10, 100, 1000])

# Applying log transformation
X_log = np.log(X)

print(X_log)
```

image.png

```python
import pandas as pd

# Example feature
X = pd.Series([22, 25, 47, 51, 19, 31, 45])

# Binning the ages into categories
bins = [0, 18, 35, 50, 100]
labels = ['Teen', 'Young Adult', 'Middle-aged', 'Senior']
X_binned = pd.cut(X, bins=bins, labels=labels)

print(X_binned)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Example feature
X = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

# Applying Min-Max scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

print(X_scaled)
```

image.png

```python
from sklearn.preprocessing import StandardScaler

# Example feature
X = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

# Applying standardization
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

print(X_standardized)
```

image.png

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Example feature with missing values
X = [[1, 2], [np.nan, 3], [7, 6], [9, np.nan]]

# Applying mean imputation
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

print(X_imputed)
```

image.png

```python
import pandas as pd

# Example categorical feature
X = pd.DataFrame({'Color': ['Red', 'Blue', 'Green', 'Blue']})

# Applying one-hot encoding
X_encoded = pd.get_dummies(X, prefix='Color')

print(X_encoded)
```
