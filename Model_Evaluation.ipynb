## Model Evaluation Metrics

When building classification models, it is crucial to evaluate their performance using appropriate metrics. Here are some of the key metrics used in classification tasks:

- **Accuracy**: The proportion of correct predictions over all predictions.
- **Precision**: The proportion of true positives out of all positive predictions.
- **Recall**: The proportion of true positives out of all actual positives.
- **F1 Score**: The harmonic mean of precision and recall, useful for imbalanced datasets.
- **Confusion Matrix**: A table that shows the counts of TP, TN, FP, and FN.
- **ROC Curve and AUC**: The ROC curve plots true positive rate vs. false positive rate, and AUC measures the overall model performance.

We will apply these metrics to evaluate k-NN and Random Forest models.

Evaluating k-NN Model with Multiple Metrics

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc

# Generating synthetic data for evaluation
X_knn_eval, y_knn_eval = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Splitting the data into training and testing sets
X_train_knn_eval, X_test_knn_eval, y_train_knn_eval, y_test_knn_eval = train_test_split(X_knn_eval, y_knn_eval, test_size=0.2, random_state=42)

# Creating the k-NN model
knn_eval = KNeighborsClassifier(n_neighbors=3)
knn_eval.fit(X_train_knn_eval, y_train_knn_eval)

# Making predictions
y_pred_knn_eval = knn_eval.predict(X_test_knn_eval)
y_prob_knn_eval = knn_eval.predict_proba(X_test_knn_eval)[:, 1]

# Evaluating the model with various metrics
accuracy_knn_eval = accuracy_score(y_test_knn_eval, y_pred_knn_eval)
precision_knn_eval = precision_score(y_test_knn_eval, y_pred_knn_eval)
recall_knn_eval = recall_score(y_test_knn_eval, y_pred_knn_eval)
f1_knn_eval = f1_score(y_test_knn_eval, y_pred_knn_eval)

# Confusion matrix
conf_matrix_knn_eval = confusion_matrix(y_test_knn_eval, y_pred_knn_eval)

# ROC curve and AUC
fpr_knn_eval, tpr_knn_eval, thresholds_knn_eval = roc_curve(y_test_knn_eval, y_prob_knn_eval)
auc_knn_eval = auc(fpr_knn_eval, tpr_knn_eval)

# Printing evaluation results
print(f"k-NN Accuracy: {accuracy_knn_eval:.2f}")
print(f"k-NN Precision: {precision_knn_eval:.2f}")
print(f"k-NN Recall: {recall_knn_eval:.2f}")
print(f"k-NN F1 Score: {f1_knn_eval:.2f}")
print(f"Confusion Matrix:\n{conf_matrix_knn_eval}")

# Plotting the ROC curve
plt.figure()
plt.plot(fpr_knn_eval, tpr_knn_eval, color='blue', lw=2, label=f'ROC curve (AUC = {auc_knn_eval:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for k-NN')
plt.legend(loc="lower right")
plt.show()
```

Evaluating Random Forest Model with Multiple Metrics

```python
from sklearn.ensemble import RandomForestClassifier

# Creating synthetic data for Random Forest evaluation
X_rf_eval, y_rf_eval = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Splitting the data into training and testing sets
X_train_rf_eval, X_test_rf_eval, y_train_rf_eval, y_test_rf_eval = train_test_split(X_rf_eval, y_rf_eval, test_size=0.2, random_state=42)

# Creating the Random Forest model
rf_eval = RandomForestClassifier(n_estimators=100, random_state=42)
rf_eval.fit(X_train_rf_eval, y_train_rf_eval)

# Making predictions
y_pred_rf_eval = rf_eval.predict(X_test_rf_eval)
y_prob_rf_eval = rf_eval.predict_proba(X_test_rf_eval)[:, 1]

# Evaluating the model with various metrics
accuracy_rf_eval = accuracy_score(y_test_rf_eval, y_pred_rf_eval)
precision_rf_eval = precision_score(y_test_rf_eval, y_pred_rf_eval)
recall_rf_eval = recall_score(y_test_rf_eval, y_pred_rf_eval)
f1_rf_eval = f1_score(y_test_rf_eval, y_pred_rf_eval)

# Confusion matrix
conf_matrix_rf_eval = confusion_matrix(y_test_rf_eval, y_pred_rf_eval)

# ROC curve and AUC
fpr_rf_eval, tpr_rf_eval, thresholds_rf_eval = roc_curve(y_test_rf_eval, y_prob_rf_eval)
auc_rf_eval = auc(fpr_rf_eval, tpr_rf_eval)

# Printing evaluation results
print(f"Random Forest Accuracy: {accuracy_rf_eval:.2f}")
print(f"Random Forest Precision: {precision_rf_eval:.2f}")
print(f"Random Forest Recall: {recall_rf_eval:.2f}")
print(f"Random Forest F1 Score: {f1_rf_eval:.2f}")
print(f"Confusion Matrix:\n{conf_matrix_rf_eval}")

# Plotting the ROC curve
plt.figure()
plt.plot(fpr_rf_eval, tpr_rf_eval, color='green', lw=2, label=f'ROC curve (AUC = {auc_rf_eval:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Random Forest')
plt.legend(loc="lower right")
plt.show()
```

### Model Evaluation Conclusion

- **Accuracy**: Provides a quick measure of performance but can be misleading for imbalanced datasets.
- **Precision**: Measures how many of the predicted positives were actually positive.
- **Recall**: Measures how many of the actual positives were identified correctly.
- **F1 Score**: A balanced measure of precision and recall, useful for imbalanced datasets.
- **Confusion Matrix**: Gives detailed insights into the types of errors the model makes.
- **ROC Curve & AUC**: The ROC curve shows how the model performs across different thresholds, and AUC provides an overall performance measure.

Both **k-NN** and **Random Forest** show good performance, but Random Forest tends to have a more stable decision boundary and higher performance metrics, especially as the data complexity increases.
