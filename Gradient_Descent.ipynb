#Gradient Descent

Gradient Descent is an optimization algorithm used to minimize a cost function in machine learning and deep learning models. It's widely used in training models like linear regression, logistic regression, and neural networks.

The idea is to iteratively update the model parameters (weights) to move towards the direction where the cost (error) is minimized. This is done by calculating the gradient (slope) of the cost function and taking small steps in the direction of the steepest descent.

## Types of Gradient Descent:

1. **Batch Gradient Descent**:
   - Updates parameters after computing the gradients over the entire dataset.
   - **Advantage**: Stable updates, but can be slow with large datasets.
   - **Disadvantage**: High computational cost for large datasets.

2. **Stochastic Gradient Descent (SGD)**:
   - Updates parameters after each training example.
   - **Advantage**: Faster for large datasets.
   - **Disadvantage**: More noisy updates, which can lead to instability.

3. **Mini-batch Gradient Descent**:
   - Compromise between Batch and SGD, updating parameters after processing a small batch of examples.
   - **Advantage**: More efficient than Batch, less noisy than SGD.

### Key Points:
- **Learning Rate (Î±)**: Controls the step size in each iteration. A higher learning rate can cause the model to overshoot, while a very low learning rate results in slow convergence.
- **Convergence**: Gradient descent converges when the updates become very small, indicating the cost function is minimized.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for linear regression (y = 4 + 3x + noise)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Adding bias (intercept) term to the feature matrix
X_b = np.c_[np.ones((100, 1)), X]

# Parameters: learning rate and number of iterations
learning_rate = 0.1
n_iterations = 1000
m = len(X_b)

# Initialize theta (weights) randomly
theta = np.random.randn(2, 1)

# Function to compute the cost (MSE)
def compute_cost(X_b, y, theta):
    predictions = X_b.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Gradient Descent implementation
def gradient_descent(X_b, y, theta, learning_rate, n_iterations):
    cost_history = []
    for iteration in range(n_iterations):
        gradients = (1 / m) * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - learning_rate * gradients
        cost = compute_cost(X_b, y, theta)
        cost_history.append(cost)
    return theta, cost_history

# Running gradient descent
theta_best, cost_history = gradient_descent(X_b, y, theta, learning_rate, n_iterations)

# Final parameters (theta)
print("Best theta (parameters):", theta_best)

# Plotting the cost function
plt.plot(cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Function Over Iterations")
plt.show()

# Visualizing the regression line
plt.scatter(X, y, label="Data")
plt.plot(X, X_b.dot(theta_best), color="red", label="Best Fit Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
